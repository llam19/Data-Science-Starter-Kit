## Descriptive Statistics

```python
# TODO: Load the dataset and compute mean, median, mode, std, IQR
import pandas as pd
import numpy as np
df = pd.read_csv('../data/ACCIDENTS_GU_BCN_2013.csv')
victims = df['Nï¿½mero de vï¿½ctimes']
print('Mean:', victims.mean())
print('Median:', victims.median())
print('Mode:', victims.mode().iloc[0])
print('Standard deviation:', victims.std())
print('Interquartile Range:', victims.quantile(0.75) - victims.quantile(0.25))
```

## Visualizations (Histogram, KDE, Boxplot, Heatmap)

```python
# TODO: Plot histogram, KDE, boxplot, and correlation heatmap
# Histogram
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../data/ACCIDENTS_GU_BCN_2013.csv')
victims = df['Nï¿½mero de vï¿½ctimes'].dropna()
plt.hist(victims, bins=15)
plt.title('Histogram')
plt.show()

# KDE
victims.plot(kind='kde', title='Kernel Density Estimate')
plt.grid(True)
plt.show

# Boxplot
plt.figure()
plt.boxplot(victims)
plt.title('Boxplot')
plt.grid(True)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
df_numeric = df.select_dtypes(include=['number'])
correlation_matrix = df_numeric.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidth=0.5)
plt.title('Correlation Heatmap')
plt.show

```

## Bootstrap Confidence Interval

```python
# TODO: Bootstrap sample mean and CI (e.g., for 'NÃºmero de vÃ­ctimes')
import numpy as np
import pandas as pd
df = pd.read_csv('../data/ACCIDENTS_GU_BCN_2013.csv')
victims = df['Nï¿½mero de vï¿½ctimes'].dropna()
boot_means = [np.mean(np.random.choice(victims, size=len(victims), replace=True)) for _ in range(1000)]
ci_upper = np.percentile(boot_means, 97.5)
ci_lower = np.percentile(boot_means, 2.5)
print(f'95% Confidence Interval for mean victims: ({ci_lower:.2f}, {ci_upper:.2f})')
```

## Hypothesis Testing

```python
# TODO: Perform a t-test (e.g., is mean != 1?) and interpret the result
import pandas as pd
from scipy import stats
df = pd.read_csv('../data/ACCIDENTS_GU_BCN_2013.csv')
victims = df['Nï¿½mero de vï¿½ctimes'].dropna()
test_statistic, p_value = stats.ttest_1samp(victims, popmean=1)
print(f"t-statistic: {test_statistic:.3f}")
print(f"p-value: {p_value:.4f}")
if p_value < 0.05:
    print("Result: Refect null hypothesis")
else:
    print("Result: Fail to reject null hypothesis")
```
